**Цель**: разработать программное обеспечение, моделирующее процессы взаимной диффузии, наблюдаемые при эффектах Френкеля и Киркендалла и осуществляющее визуализацию образование пор и полостей в сферических частицах. Соотнести полученные результаты моделирования с экспериментальными результатами, полученными в научной группе УНЛ "Наноматериалы".

## Перечень допущений компьютерной модели

Перечень допущений компьютерной модели
1) Движение вакансий Pb и Se осуществляется только по собственным подрешёткам, молекулы кислорода могут двигаться по любым вакансиям. Данный подход основан на том, что атомы имеют различный характерный размер.
2) По теории Френкеля источником вакансий является поверхность. Вакансии, продвигаясь внутрь структуры, вытесняют к поверхности атомы Pb.
3) Так как по ходу моделирования атомы Pb устремляются к поверхности, то они начинают препятствовать атомам кислорода проникать внутрь кристаллической структуры, и на поверхности образуется аморфный слой PbO. Поскольку реакция записывается в виде

2PbSe + O2 = 2PbO + 2Se!

то логично, что атомы Se, уходя из структуры к поверхности, испаряются
4) Кристаллическая структура строится на основании, что каждый атом имеет соседей с 8 сторон (кроме поверхности и пор), а аморфная структура предполагает отсутствие строгой упорядоченности (атомы имеют соседей не со всех 8 сторон)

2. Предлагаемый алгоритм работы программы
1. Объявляется 2 двумерных массива. Первый массив хранит местоположение частиц в сетке, второй – хранит значения энергий частиц. Распределение энергии между частицами задаётся пользователем (Гауссово, нормальное).
