## Описание эффекта Киркендалла

## Описание экспериментальных наблюдений

В поликристаллических фотоприемниках на основе полупроводников $A^\text{II}B^\text{VI}$ (например, PbSe) фоточувствительность материала в существенной мере определяется структурой зерен и наличием оксида, в частности PbO и PbSeO$_3$.  В статье [1] для непосредственного изучения внутреннего строения зерен поликристалических образцов полупроводниковых сенсоров были проведены эксперименты по механическому сечению зерен зондом атомно-силового микроскопа (Рис. 1).

Для проведения силовой литографии был использован кремниевый зонд для полуконтактного режима ($R_\text{закр} ≥ 10$ нм) с константой жесткости $k$ в диапазоне 5.5-10$ Н/м. Результаты эксперимента по механическому сечению зерна зондом приведены на Рис. 1. По данным микроскопии толщина оксидного покрытия зерен составляет ~10–12 нм.

![Изображение сечения зерна](images/afm.png)

Рис. 1. Изображение топографии разреза зерна (а) и того же участка поверхности по данным латерально-силовой микроскопии (б)

Относительно структуры зерна можно предположить пористый характер структуры в приповерхностной области. Объяснение процессу образования пор было дано в работе [2]. В основе рассмотрения лежит модель, близкая к используемой для описания классического эффекта Киркендалла – существенную роль играет различие в значениях коэффициентов диффузии взаимодействующих элементов.

Пусть вначале имеется зерно фоточувствительного материала состава PbSe с избытком свинца, соответствующим n-типу проводимости. Пусть это зерно подвергается окислению и на поверхности образуется оксид. Из-за того, что значения коэффициента диффузии атомов свинца выше значений коэффициента диффузии кислорода, рост оксидных фаз происходит на внешней стороне окисляемого зерна.

Задача моделирования состоит в исследовании процесса образования пор и слоя оксида при помощи атомистической модели.

## Предлагаемый алгоритм работы программы

Задачу предлагается решать подредством представления двумерного среза кристаллической решетки в виде двумерного массива, преобразующегося по следующему алгоритму (соответствует Рис. 2). В процессе проработки алгоритма модель будет уточняться.

Задается двумерный массив целых чисел, в котором каждое число соответствует различному типу ионов, например:
    - 0 - Вакансия
    - 1 - Pb
    - 2 - Se
    - 3 - O.
    - 4 - Кислород, связавшийся со свинцом обозначается.
    - 5 - Вакансия, вошедшая в пору стается неподвижной.

![](draft_model.png)
Рис. 2. Визуализация двумерного массива, соответствующего началу работы модели/

Расположение и количество чисел в массиве обуславливаются следующими правилами:
- Построение начинается с ионов свинца [1] и селена [2], располагающихся в соответствии со структурой кристаллической решетки (PbSe соответствует структуре NaCl). Для удобства визуализации имеет смысл сначала моделировать небольшой размера масива NxN, например, 25х25.
- Количество вакансий [0] задается, исходя из типичного отклонения от стехиометрии состава и раскидываются по массиву случайным образом в местах тех атомов, которых имеется в недостатке, например, атомов селена [2]. Доля вакансий для простейшей модели задается в виде константы, например, 1%
- В дополнение к массиву создается строка, соответствующая среде, с которой контактирует кристалл. По теории Френкеля источником вакансий является поверхность. Вакансии, продвигаясь внутрь структуры, вытесняют к поверхности атомы металла и халькогена. То есть поверхность служит источником вакансий [0] (образование деффектов по Шоттки) и ионов кислорода [3], таким образом эта строка массива состоит из [0] из [3]. Отношение  числа вакансий ("вакуума") [0] и кислорода [3] зависит от давления газа – чем оно больше, тем больше [3]. Вначале это модельное отношение задается случайно. Исходно "прилегающее пространство" первой строки массива состоит из [0] и [3], но по мере движения атомов происходит образование оксида. Если в эту строку попадает число, отличное от [0] и [3] то "вверх" наращивается еще одна строка массива. 
- Рассматриваются только движение вакансий [0] и ионов кислорода [3] (перемещение ионов селена или свинца равнозначно перемещению вакансий):
    - В каждой итерации атомы кислорода [3] перемещается произвольно на одно место по горизонтали, вертикали или диагонали. Если атом [3] касается [1], атом [3] перестает двигаться и переобозначается как [4] (образуется оксид свинца), вакансия [0] может распространяться как по [1], так и по [4]. При перемещении [0] и [3], меняются местами с тем ионом, который был ими замещен
    - В движение приводятся только те вакансии [0] и ионы кислорода [3], которые имеют в качестве хотя бы одного из соседей атом свинца [1], селена [2] или связанного кислорода [4], то есть вакансии перемещающиеся по твердой фазе. В каждой итерации вакансия [0] каждый раз перемещается на 2 места по горизонтали, вертикали или диагонали, соответствуя своей подрешетке. Правила превращения вакансии в пору: 1) Если вакансия [0] касается вакансия  [0] -> оба [0] становятся [5] - сливаются в пору и перестают двигаться, 2) Если вакансия [0] касается поры [5] -> [0] становится [5]
- Конец работы программы задается количеством производимых иттераций.	


    
Усложняющие уточнения, пока не принимаемые в расчет:
- Температура влияет на скорость движения вакансий, скорость движения кислорода (то, как далеко перебрасывается [0] и [3])
